```python
import torch
model_file = torch.load("full_trained_gpt_nparams-21072400_steps-100.pt")
model_file
```

```python
state_dict = model_file["model"]
state_dict
```

```python
from transformers import AutoModelForCausalLM, GPT2Config, set_seed

model_config = GPT2Config(
    vocab_size=50257,
    n_positions=20,
    n_embd=200,
    n_layer=2,
    n_head=10,
    n_inner=None,
    activation_function="gelu_new",
    resid_pdrop=0.1,
    embd_pdrop=0.1,
    attn_pdrop=0.1,
    layer_norm_epsilon=1e-05,
    initializer_range=0.02,
    summary_type="cls_index",
    summary_use_proj=True,
    summary_activation=None,
    summary_proj_to_labels=True,
    summary_first_dropout=True,
    # gradient_checkpointing=args.gradient_checkpointing > 0,
    use_cache=False,
    bos_token_id=50256,
    eos_token_id=50256,
    return_dict=True,
)
```

```python
model = AutoModelForCausalLM.from_config(model_config)
```

```python
model.load_state_dict(state_dict)
```

```python
model.eval()
```
